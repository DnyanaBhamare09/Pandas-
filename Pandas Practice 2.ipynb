```python
url3 = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv"
```

```python
df3 = pd.read_csv(url3)
```

```python
df3.head(5)
```

```python
# Convert the type of the column Year to datetime 64
```

```python
df3.Year = pd.to_datetime(df3.Year, format = "%Y")
```

```python
df3.info()
```

```python
# Set the Year column as the index of the dataframe
```

```python
df3 = df3.set_index("Year", drop = True)
```

```python
df3.head()
```

```python
# Group the year by decades and sum the values
```

```python
# Uses resample to sum each decade
df3 = df3.resample("10AS").sum()
```

```python
# Uses resample to get the max value only for the "Population" column
population = df3['Population'].resample('10AS').max()
```

```python
# Updating the "Population" column
df3['Population'] = population
```
